```python
import pandas as pd
```

```python
df = pd.read_csv("kc_house_data.csv")
```

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

```python
df.pivot_table(index='waterfront',
               columns='view',
               values='id',
               aggfunc='count',
               fill_value=0)
```

```python
pd.crosstab(index=df['floors'],
            columns=df['bedrooms'])
```

```python
df.pivot_table(index='condition', columns='grade', values='price', aggfunc='median', fill_value=0)
```
